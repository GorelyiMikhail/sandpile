```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt, log10
```

```python
ends = [10**3, 10**4, 10**5, 10**6]
sizes = [64, 128, 256, 512, 1024, 2048]
```

```python
degs_100 = []

fig, ax = plt.subplots()

for size in sizes:
    distribution = open(f'distributions/distribution{size}_100.out', "r")
    y = list(map(int, distribution.read().split()))
    start, cur_size = 0, 1
    new_y = []
    bin_mid = []
    while start + cur_size < len(y):
        new_y.append(sum(y[start:start + round(cur_size)]))
        bin_mid.append(sqrt(start * (start + round(cur_size))))
        start += round(cur_size)
        cur_size *= 1.2
    ax.plot(bin_mid, new_y, "-o", label=f'{size} 0.001')

    start = 0
    all_angles = [[], [], [], []]
    for i in range(len(new_y)):
        if bin_mid[i] >= 200 and start == 0:
            start = i

        for j in range(len(ends)):
            if start != 0 and i != start and bin_mid[i] <= ends[j] and new_y[i] > 0:
                all_angles[j].append((log10(new_y[start]) - log10(new_y[i])) / (log10(bin_mid[start]) - log10(bin_mid[i])))

    angles = []
    for i in range(len(all_angles)):
        sum_deg = 0
        for j in range(len(all_angles[i])):
            sum_deg += all_angles[i][j]
        angles.append(sum_deg / len(all_angles[i]))
    degs_100.append(angles)

ax.set_xlabel('Sizes of avalanche united into bins, log scale', color='red')
ax.set_ylabel('Amount of avalanches, log scale', color='red')
ax.set_title("Amount of avalanches of each size", color='red')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylim([10 ** 3, 10 ** 5])
ax.set_xlim([1, 10 ** 7])
ax.tick_params(color='red', labelcolor='red')
fig.set_size_inches(18.5, 10.5)
plt.legend()
plt.show()
fig.savefig('distributions0.01.png')
```

```python
fig, ax = plt.subplots()
x = np.arange(len(sizes))

for i in range(len(ends)):
    values = []

    for j in range(len(sizes)):
        values.append(degs_100[j][i])
    
    ax.plot(x, values, "-o", label=f'[200, 10^{int(log10(ends[i]))}]')

ax.set_xlabel('Grid size', color='red')
ax.set_ylabel('Angle', color='red')
ax.set_title('Angles of linear part for different grid sizes', color='red')
ax.tick_params(color='red', labelcolor='red')
plt.xticks(x, sizes)
fig.set_size_inches(18.5, 10.5)
plt.legend()
plt.show()
fig.savefig('angles0.01.png')
```
